```python
# NanoDet-Plus-m (1.0x) - Optimized for Raspberry Pi 3B+
# Workflows: Training 320x320
save_dir: ./workspace/nanodet_m_1x_480
model:
  weight_averager:
    name: ExpMovingAverager
    decay: 0.9998
  arch:
    name: NanoDetPlus
    detach_epoch: 10
    backbone:
      name: ShuffleNetV2
      model_size: 1.0x           # <--- CHANGED: Smaller backbone for Pi 3B+
      out_stages: [2, 3, 4]
      activation: LeakyReLU
    fpn:
      name: GhostPAN
      in_channels: [116, 232, 464] # <--- CHANGED: Matches 1.0x backbone
      out_channels: 96             # <--- CHANGED: Reduced from 128 for speed
      kernel_size: 5
      num_extra_level: 1
      use_depthwise: True
      activation: LeakyReLU
    head:
      name: NanoDetPlusHead
      num_classes: 3             # <--- Make sure your Max Class ID is 2 (0,1,2)
      input_channel: 96          # <--- CHANGED: Matches FPN
      feat_channels: 96          # <--- CHANGED: Matches FPN
      stacked_convs: 2
      kernel_size: 5
      strides: [8, 16, 32, 64]
      activation: LeakyReLU
      reg_max: 7
      loss:
        loss_qfl:
          name: QualityFocalLoss
          use_sigmoid: True
          beta: 2.0
          loss_weight: 1.0
        loss_dfl:
          name: DistributionFocalLoss
          loss_weight: 0.25
        loss_bbox:
          name: GIoULoss
          loss_weight: 2.0
    aux_head:
      name: SimpleConvHead
      num_classes: 3             # <--- Must match 'head' num_classes
      input_channel: 192         # <--- CHANGED: 96 * 2 (Always double the head)
      feat_channels: 192         # <--- CHANGED: Matches input
      stacked_convs: 4
      strides: [8, 16, 32, 64]
      activation: LeakyReLU
      reg_max: 7

class_names: &class_names ['adult', 'child', 'pet']

data:
  train:
    name: CocoDataset
    img_path: /content/dataset/train
    ann_path: /content/dataset/train/_annotations.coco.json
    input_size: [480,480]          # <--- CHANGED: 320 is best for Pi 3B+
    keep_ratio: False
    pipeline:
      perspective: 0.0
      scale: [0.6, 1.4]
      stretch: [[0.8, 1.2], [0.8, 1.2]]
      rotation: 0
      shear: 0
      translate: 0.2
      flip: 0.5
      brightness: 0.2
      contrast: [0.6, 1.4]
      normalize: [[103.53, 116.28, 123.675], [57.375, 57.12, 58.395]]
  val:
    name: CocoDataset
    img_path: /content/dataset/valid
    ann_path: /content/dataset/valid/_annotations.coco.json
    input_size: [480,480]          # <--- CHANGED: Match train size
    keep_ratio: False
    pipeline:
      normalize: [[103.53, 116.28, 123.675], [57.375, 57.12, 58.395]]
  test:
    name: CocoDataset
    img_path: /content/dataset/test
    ann_path: /content/dataset/test/_annotations.coco.json
    input_size: [480,480]        # <--- CHANGED: Match train size
    keep_ratio: False
    pipeline:
      normalize: [[103.53, 116.28, 123.675], [57.375, 57.12, 58.395]]

device:
  gpu_ids: [0]
  workers_per_gpu: 8
  batchsize_per_gpu: 32   # If you get "CUDA Out of Memory", change this to 16

schedule:
  optimizer:
    name: AdamW
    lr: 0.001
    weight_decay: 0.05
  warmup:
    name: linear
    steps: 500
    ratio: 0.0001
  total_epochs: 250
  lr_schedule:
    name: CosineAnnealingLR
    T_max: 150
    eta_min: 0.00005
  val_intervals: 10

grad_clip: 35
evaluator:
  name: CocoDetectionEvaluator
  save_key: mAP
log:
  interval: 10
```
